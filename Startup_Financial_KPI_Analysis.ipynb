```python
import pandas as pd

# Read the correct sheet
df = pd.read_excel(
    "Startup_Financial_KPI_Dataset.xlsx",
    sheet_name="Startup_Financials"
)

# Check data
df.head()
```

```python
# Dataset information
print(df.info())

# Statistical summary
df.describe()
```

```python
# Check missing values
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.plot(df["Month"], df["Revenue"], marker="o", linewidth=3)

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df["Month"], df["Expenses"],
         marker="o",
         linewidth=3,
         color="red")

plt.title("Monthly Expenses Trend")
plt.xlabel("Month")
plt.ylabel("Expenses")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df["Month"], df["Profit"],
         marker="o",
         linewidth=3,
         color="green")

plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df["Month"], df["Active_Customers"],
         marker="o",
         linewidth=3,
         color="purple")

plt.title("Active Customers Growth")
plt.xlabel("Month")
plt.ylabel("Active Customers")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df["Month"], df["LTV"],
         marker="o",
         linewidth=3,
         label="LTV")

plt.plot(df["Month"], df["CAC"],
         marker="s",
         linewidth=3,
         label="CAC")

plt.title("LTV vs CAC Comparison")
plt.xlabel("Month")
plt.ylabel("Value")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.bar(df["Month"], df["Profit Margin %"])

plt.title("Monthly Profit Margin")
plt.xlabel("Month")
plt.ylabel("Profit Margin (%)")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df["Month"], df["Burn Rate"],
         marker="o",
         linewidth=3,
         color="orange")

plt.title("Monthly Burn Rate")
plt.xlabel("Month")
plt.ylabel("Burn Rate")
plt.grid(True)

plt.show()
```

Business Insights

1. Revenue increased consistently from January to December.
2. Profit also showed steady monthly growth.
3. Active customers increased significantly throughout the year.
4. Marketing investment resulted in customer growth.
5. LTV remained much higher than CAC, indicating efficient customer acquisition.
6. Burn rate increased as the startup scaled its operations.
7. Overall financial performance remained healthy and sustainable.
